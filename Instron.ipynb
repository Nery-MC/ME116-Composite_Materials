```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Load Data
Instron_Path = '/Users/nery.matias_calmo/Desktop/ReFormatted.csv'
Instron_Data = pd.read_csv(Instron_Path, header=0)
print(f'Instron Data Shape: {Instron_Data.shape}')
```

```python
# Extract units for each variable (assuming second row contains units)
Units = Instron_Data.iloc[0,0:].values  

Instron_Data.iloc[1:, 0:] = Instron_Data.iloc[1:, 0:].apply(pd.to_numeric, errors='coerce')

print(Units)
```

```python
Sample_Ranges = {'Sample 1': (1, 5981), 'Sample 2': (5982, 14557), 'Sample 3': (14558, 21853), 'Sample 4': (21854, 29968), 'Sample 5': (29969, 42842)}
```

```python
# Plot Displacement vs. Time
plt.figure(figsize = (10, 6))

for Sample_Name, (Start, End) in Sample_Ranges.items(): 
    Time = Instron_Data.iloc[Start:End, 0].dropna().values.flatten()
    Displacement = Instron_Data.iloc[Start:End, 1].dropna().values.flatten()
    
    plt.plot(Time, Displacement, label = Sample_Name, linewidth = 3)

plt.title('Displacement Time Response')
plt.xlabel(f'Time {Units[0]}')  
plt.ylabel(f'Displacement {Units[1]}')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Plot Force vs. Time
plt.figure(figsize = (10, 6))

for Sample_Name, (Start, End) in Sample_Ranges.items(): 
    Time = Instron_Data.iloc[Start:End, 0].dropna().values.flatten()
    Force = Instron_Data.iloc[Start:End, 2].dropna().values.flatten()
    
    plt.plot(Time, Force, label = Sample_Name, linewidth = 3)

plt.title('Force Time Response')
plt.xlabel(f'Time {Units[0]}')  
plt.ylabel(f'Force {Units[2]}')
plt.legend()
plt.grid(True)
```

```python
# Plot Strain vs. Stress
plt.figure(figsize = (10, 6))

for Sample_Name, (Start, End) in Sample_Ranges.items(): 
    Tensile_Strain = Instron_Data.iloc[Start:End, 3].dropna().values.flatten()
    Tensile_Stress = Instron_Data.iloc[Start:End, 4].dropna().values.flatten()
    
    plt.plot(Tensile_Strain, Tensile_Stress, label = Sample_Name, linewidth = 3)

plt.title('Stress Strain Cruve')
plt.xlabel(f'Tensile Strain {Units[3]}')  
plt.ylabel(f'Tensile Stress {Units[4]}')
plt.legend()
plt.grid(True)
```

```python
# Plot Force vs. Displacement
plt.figure(figsize = (10, 6))

for Sample_Name, (Start, End) in Sample_Ranges.items(): 
    Force = Instron_Data.iloc[Start:End, 2].dropna().values.flatten()
    Displacement = Instron_Data.iloc[Start:End, 1].dropna().values.flatten()
    
    plt.plot(Displacement, Force, label = Sample_Name, linewidth = 3)

plt.title('Force Displacement Cruve')
plt.xlabel(f'Displacement {Units[1]}')  
plt.ylabel(f'Force {Units[2]}')
plt.legend()
plt.grid(True)
```
